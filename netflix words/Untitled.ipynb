```python
import sqlite3
```

```python
conn = sqlite3.connect('db.db')
```

```python
def make_table(name, col_names, col_types):
    c = conn.cursor()
    columns = ", ".join([str(x[0]) + " " + str(x[1]) for x in zip(col_names, col_types)])
    table_creation_query = f"CREATE TABLE {name} ({columns})"
    c.execute(table_creation_query)
```

```python
def fill_table(name, data):
    c = conn.cursor()
    values_placeholder = ','.join(['?'] * len(data[0]))
    c.executemany(f'INSERT INTO {name} VALUES ({values_placeholder})', data)
```

```python
make_table('jlpt_n1_words', ['kanji', 'pron', 'meaning'], ['text', 'text', 'text'])
fill_table('jlpt_n1_words', jlpt_words)
```

```python
conn.commit()
```

```python
jlpt_n5 = open("jlpt_5.txt", "r", encoding="utf-8").readlines()
jlpt_n5_csv = []
for i in range(int(len(jlpt_n5)/3)):
    jlpt_n5_csv.append([jlpt_n5[3*i], jlpt_n5[3*i + 1], jlpt_n5[3*i + 2]])
    
jlpt_n5_csv = [[y.strip() for y in x] for x in jlpt_n5_csv]
make_table('jlpt_n5_words', ['kanji', 'pron', 'meaning'], ['text', 'text', 'text'])
```

```python
fill_table('jlpt_n5_words', jlpt_n5_csv)
conn.commit()
```
